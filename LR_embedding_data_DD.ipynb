```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import re
from scipy import spatial
from google.colab import drive
drive.mount('/content/drive')

%cd /content/drive/My \Drive/Datos/BLEU

DD_reference_dataset = pd.read_csv("DAILYD_translation_en2es.csv")
DD_translation_dataset = pd.read_csv("DAILYD_translation_es2en.csv")

DD_reference = DD_reference_dataset['SEG']
DD_translation = DD_reference_dataset['translation']
DD_candidate = DD_translation_dataset['translation']
n_DD=len(DD_reference)
print(len(DD_reference))
```

```python
model_1_ing_ing=SentenceTransformer('paraphrase-TinyBERT-L6-v2')
model_2_ing_esp=SentenceTransformer('distiluse-base-multilingual-cased-v1')
model_3_ing_esp=SentenceTransformer('paraphrase-xlm-r-multilingual-v1')

data_matrix=[]

for x in range(len(DD_reference)):
  ref = re.sub(r'\s([?.!",’](?:\s|$))', r'\1', DD_reference[x])
  cand = re.sub(r'\s([?.!"](?:\s|$))', r'\1', DD_candidate[x])
  trans = re.sub(r'\s([?.!"](?:\s|$))', r'\1', DD_translation[x])

  emb_trans = [model_2_ing_esp.encode(trans),model_3_ing_esp.encode(trans)]
  emb_ref = [model_1_ing_ing.encode(ref),model_2_ing_esp.encode(ref),model_3_ing_esp.encode(ref)]
  emb_cand = [model_1_ing_ing.encode(cand)]

  cos_results=[1-spatial.distance.cosine(emb_ref[0],emb_cand[0]),1-spatial.distance.cosine(emb_ref[1],emb_trans[0]),1-spatial.distance.cosine(emb_ref[2],emb_trans[1])]
  
  print(x,cos_results)

  data_matrix.append(cos_results)

```

```python
df=pd.DataFrame(data_matrix)
df.columns = ['modelo_1_mono','modelo_2_multi','modelo_3_multi']
df.to_excel(excel_writer = "LR_data.xlsx")
```
